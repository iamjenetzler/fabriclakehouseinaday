# Shortcut keys

The keyboard does different things depending on which mode the notebook cell is in. Fabric notebooks support the following two modes for a given code cell: <b>Command mode and Edit mode</b>.

## Command mode

A cell is in Command mode when there's no text cursor prompting you to type. When a cell is in Command mode, you can edit the notebook as a whole but not type into individual cells. Enter Command mode by pressing ESC or using the mouse to select outside of a cell's editor area.

<img src="https://learn.microsoft.com/en-us/fabric/data-engineering/media/author-execute-notebook/cell-command-mode.png#lightbox">

### Some of the most commonly used shortcuts in command mode are:

<b>Shift+Enter</b>:         Executes the current cell and moves to next cell below.

<b>Shift+Enter</b>:         Executes the current cell and moves to next cell below.

<b>Ctrl+Enter</b>:          Executes the highlighted lines of code in a cell

<b>Down</b>:                Select cell below

<b>Up</b>:                  Select cell above

<b>K</b>:                   Select previous cell	

<b>J</b>:                   Select next cell

<b>A</b>:                   Insert cell above

<b>B</b>:                   Insert cell below	

<b>Shift+D</b>:             Delete selected cells	

<b>Enter</b>:               Switch to edit mode

## Edit mode

Edit mode can be indicated from a text cursor that prompting you to type in the editor area. When a cell is in Edit mode, you can type into the cell. Enter Edit mode by pressing Enter or using the mouse to select on a cell's editor area.

<img src="https://learn.microsoft.com/en-us/fabric/data-engineering/media/author-execute-notebook/cell-edit-mode.png#lightbox">

### Some of the most commonly used shortcuts in Edit mode are:

<b>Up</b>:		Move cursor up

<b>Down</b>:		Move cursor down

<b>Ctrl + Z</b>:		Undo

<b>Ctrl + Y</b>:		Redo

<b>Ctrl + /</b>:		Comment or Uncomment

<b>Ctrl + Backspace</b>:		Delete word before

<b>Ctrl + Delete</b>:		Delete word after

<b>Ctrl + Home</b>:		Go to cell start

<b>Ctrl + End</b>:		Go to cell end

<b>Ctrl + Left</b>:		Go one word left

<b>Ctrl + Right</b>:		Go one word right

<b>Ctrl + A</b>:		Select all

<b>Ctrl + ]</b>:		Indent

<b>Ctrl + [</b>:		Dedent

<b>Esc</b>:		Switch to command mode

# Magic Commands
Magic commands will allow you to override the default language for a notebook for the current cell.

<b>%%sql</b> is the magic command to use Databricks SQL for a cell

<b>%%r</b> is the magic command to use the R language for a cell

<b>%%python</b> is the magic command to use the Python language for a cell.

## Note that at the top of a notebook there is a default language.

```python
%%sql
-- I am a sql cell
```

```python
%%sparkr
# I am an r cell
```

```python
%%pyspark 
# I am a PySpark cell
```
